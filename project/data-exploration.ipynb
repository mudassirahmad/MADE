```python
import pandas as pd
from sqlalchemy import create_engine
import seaborn as sns
import matplotlib.pyplot as plt
import subprocess
import os
```

```python
tourist_data_url="https://res.cloudinary.com/dx6obccn6/raw/upload/v1701604753/tourists-data_aas2zo.csv"
tourist_states_url="https://res.cloudinary.com/dx6obccn6/raw/upload/v1704827971/tourist_areas_with_states_zwmucy.csv"
transportation_data_url="https://res.cloudinary.com/dx6obccn6/raw/upload/v1701604752/transportation_vpbfi6.csv"
```

#### Data Exploration

```python
tourist_df = pd.read_csv(tourist_data_url, encoding='latin1', sep=";")
tourist_df.head(15)
```

```python
#rading the data again but skipping first few lines to get the required data in clean format
tourist_df = pd.read_csv(tourist_data_url, encoding='latin1', sep=";", skiprows=10)
tourist_df.head(15)
```

```python
column_names = ['Year', 'Train', 'Land', 'domestic arrival', 'domestic overnight stays',
                    'domestic avg stay (days)', 'int. arrival', 'intl. overnight stays',
                    'intl. avg stay (days)', 'total arrival', 'total overnight stays',
                    'total avg stay (days)']
    #rename the columns
tourist_df.columns=column_names
tourist_df = tourist_df.iloc[:-4].reset_index(drop=True)

#dropping empty rows
tourist_df.replace('-', pd.NA, inplace=True)
tourist_df.replace('.', pd.NA, inplace=True)
clean_tourist_df = tourist_df.dropna()
clean_tourist_df['total arrival']= pd.to_numeric(clean_tourist_df['total arrival'], errors='coerce').astype('int64')
clean_tourist_df.head()
```

```python
#need to add states for each tourist place
states=pd.read_csv(tourist_states_url, encoding='latin1', sep=",")

tourist_data=(pd.merge(clean_tourist_df,states,left_on="Land",right_on="Tourist Area", how='left'))

missing_values = {
    'Oberes Maintal - Coburger Land (bis 2016)': 'Baden-Württemberg',
    'Nördlicher Schwarzwald': 'Bayern',
}

#rows_with_nan=tourist_data[tourist_data['State(s)'].isna()]

# Iterate through the custom values and fill missing 'States' accordingly
for land, state in missing_values.items():
    tourist_data.loc[tourist_data['Land'] == land, 'State(s)'] = state

tourist_data=tourist_data.drop('Tourist Area', axis=1)
tourist_data.head()
```

```python
transport_df = pd.read_csv(transportation_data_url, encoding='latin1', sep=";", skiprows=7)
transport_df.head()
  
```

```python
column_names=['Year', 'Company Type', 'State(s)','Companies', 'Person Transported', 
                'Transportation Per KM', 'Mileage', 'Transportation Offer', 
                'Revenue from Local Transport (EUR)']

#ensuring LongIntegerColumn are displayed as full integers
pd.set_option('display.float_format', lambda x: '{:.0f}'.format(x))

#rename the columns
transport_df.columns=column_names
clean_transport_df = transport_df.iloc[:-3].reset_index(drop=True)
clean_transport_df.drop(["Revenue from Local Transport (EUR)","Mileage", "Transportation Offer"], axis=1, inplace=True)
clean_transport_df.head()
```
